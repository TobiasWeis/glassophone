```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# abstände und durchmesser für gläser
r_glass_w = 80/2 #wein: 70/2 # ikea: 86/2 # mm
r_glass_s = 65.5/2
r_hammer = 60 #mm
```

```python
def point_from_angle(angle_deg, r, cx=0,cy=0):
    a = np.radians(angle_deg)
    x = cx + r * np.cos(a)
    y = cy + r * np.sin(a)
    
    return x,y

fig, ax = plt.subplots(figsize=(10,10))

for i in range(0,360-3*45,45):
    p = point_from_angle(i, r_hammer+r_glass_w)
    print("Angle:",i)
    print("Point:",p)
    
    ax.add_patch(plt.Circle(p, r_glass_w, color='blue'))

for i in range(360-3*45,360,45):
    p = point_from_angle(i, r_hammer+r_glass_s)
    print("Angle:",i)
    print("Point:",p)
    ax.add_patch(plt.Circle(p, r_glass_s, color='yellow'))
    
ax.add_patch(plt.Circle((0,0), r_hammer, color='green'))    
    
plt.axis('equal')    
plt.show()
```
